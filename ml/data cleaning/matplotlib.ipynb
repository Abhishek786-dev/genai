```python
# Line Graph
price = [48,54,67,49,47,45]
year = [2015,2016,2017,2018,2019,2020]
```

```python
plt.plot(year, price, color='green', linestyle='solid', marker='o', label='Line 1')
plt.xlabel("X-axis Years")
plt.ylabel("Y-axis Price")
plt.title("Iphone Prices by Years")
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5) # Displays the labels defined above
plt.show()
```

```python
# Now we will try to plot  2-line chart in a single plot
# Data for the first line
x1 = np.array([0, 1, 2, 3, 4])
y1 = np.array([0, 2, 4, 6, 8])

# Data for the second line
x2 = np.array([0, 1, 2, 3, 4])
y2 = np.array([0, 3, 6, 9, 12])

# Plot the first line (green solid line with circle markers)
plt.plot(x1, y1, color='green', linestyle='solid', marker='o', label='Line 1')

# Plot the second line (red dashed line)
plt.plot(x2, y2, 'r--', label='Line 2') # Using format string 'r--'

# Add title, labels, and legend
plt.title("Multiple Lines with Customization")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend() # Displays the labels defined above

# Add grid lines for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.show()
```

```python
# Line Chart with Annotations
price = [48,54,67,49,47,45]
year = [2015,2016,2017,2018,2019,2020]

plt.figure(figsize=(8, 6))
plt.plot(year, price, marker='o', linestyle='-')

# Add annotations
for i, (xi, yi) in enumerate(zip(year, price)):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0, 10), ha='center')

plt.title('Line Chart with Annotations')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

plt.grid(True)
plt.show()
```

```python
#Fill the Area Between Two Lines
x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

plt.plot(x, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')

plt.fill_between(x, y, y1, color='green', alpha=0.5)
plt.show()
```

```python
# Multiple bar graph
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8)) 

IT = [12, 30, 1, 8, 22] 
ECE = [28, 6, 16, 5, 10] 
CSE = [29, 3, 24, 25, 17] 

br1 = np.arange(len(IT)) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 

plt.bar(br1, IT, color ='r', width = barWidth, edgecolor ='grey', label ='IT') 
plt.bar(br2, ECE, color ='g', width = barWidth, edgecolor ='grey', label ='ECE') 
plt.bar(br3, CSE, color ='b', width = barWidth, edgecolor ='grey', label ='CSE') 

plt.xlabel('Branch', fontweight ='bold', fontsize = 15) 
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15) 
plt.xticks([r + barWidth for r in range(len(IT))], ['2015', '2016', '2017', '2018', '2019'], rotation ='vertical') # x name will show as vertical form by default is horizontal

plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 15, 30]

# Plot the bars
plt.bar(categories, values)

# Add value labels manually
for i, v in enumerate(values):
    plt.text(i, v + 1, str(v), color='black', ha='center') # i is the x position, v + 1 is the y position

# Add plot details
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart with Values (plt.text)")

plt.show()
```

```python
# Using Subplot we are adding values on bar
# Data for 3 groups across 4 categories
categories = ['Q1', 'Q2', 'Q3', 'Q4']
product_a = [10, 15, 7, 12]
product_b = [14, 18, 11, 8]
product_c = [6, 10, 12, 16]

# Parameters for grouping
x = np.arange(len(categories))  # the label locations (numeric)
width = 0.25  # the width of a single bar

fig, ax = plt.subplots(figsize=(10, 6))

# 1. Plot all three sets of bars, offsetting each
# Product A is left of center
rects1 = ax.bar(x - width, product_a, width, label='Product A')
# Product B is centered
rects2 = ax.bar(x, product_b, width, label='Product B')
# Product C is right of center
rects3 = ax.bar(x + width, product_c, width, label='Product C')

# 2. Add labels to all three bar containers using ax.bar_label()
ax.bar_label(rects1, padding=5, fmt='{:,.0f}')
ax.bar_label(rects2, padding=5, fmt='{:,.0f}')
ax.bar_label(rects3, padding=5, fmt='{:,.0f}')

# 3. Add necessary plot details
ax.set_xlabel('Quarter')
ax.set_ylabel('Sales Volume')
ax.set_title('Quarterly Sales for Three Products')
ax.set_xticks(x, categories) # Set the custom category labels
ax.legend()

# 4. Adjust layout so labels don't get cut off
ax.margins(y=0.15) 

plt.show()
```

```python
# this is another example of sunplot where we can plot multiple plots in one single plot
df = pd.read_csv('dataset/matplotlib/batter.csv')
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
ax[0,0].scatter(df['avg'], df['strike_rate'], c='r')
ax[0,1].scatter(df['avg'], df['runs'], c='g')
ax[1,0].hist(df['avg'])
ax[1,1].hist(df['runs'])
# ax[0,0].grid(True)
fig.show()
```

```python
# one more example to use subplot
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1) # 2 rows 2 col 1st plot , if use 1,2,1 means 1 row 2 cols and 1st plot
ax1.scatter(df['avg'], df['strike_rate'], c='r')
ax2 = fig.add_subplot(2,2,2) # 2 rows 2 col 2nd plot ,
ax2.hist(df['runs'])

ax3 = fig.add_subplot(2,2,3) # 2 rows 2 col 3rd plot ,
ax3.hist(df['avg'])
```

```python
 # 3D Scatter Plots
df = pd.read_csv('dataset/matplotlib/batter.csv')
df.shape
fig = plt.figure(figsize=(10, 10))
ax = plt.subplot(projection='3d')
ax.scatter3D(df['runs'], df['avg'], df['strike_rate'])
ax.set_title('IPL Analysis')
ax.set_xlabel('Runs')
ax.set_ylabel('AVG')
ax.set_zlabel('Strike Rate')
```

```python
# Import the 3D plotting toolkit
from mpl_toolkits.mplot3d import Axes3D

# 1. Prepare data (example data for a 2x2 grid)
# The coordinates of the anchor points of the bars
xpos = [1, 1, 3, 3]
ypos = [1, 3, 1, 3]
zpos = [0, 0, 0, 0]

# The dimensions of the bars (width, depth, height)
dx = np.ones(4) # width of 1 for all bars
dy = np.ones(4) # depth of 1 for all bars
dz = [3, 2, 6, 5] # heights of each bar

# 2. Set up a figure and a 3D axes object
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# 3. Plot the bars
ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='skyblue', shade=True)

# 4. Customize the plot
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')
ax.set_zlabel('Z Axis Label (Height)')
ax.set_title('Simple 3D Bar Chart')

# 5. Display the plot
plt.show()
```
