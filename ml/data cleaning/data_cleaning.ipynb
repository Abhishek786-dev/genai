```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.DataFrame({
    'age': [25, np.nan, 35, 40],
    'salary': [50000, 60000, np.nan, 80000]
})
```

# Missing Values

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df['salary'] = df['salary'].fillna(df['salary'].median())
```

```python
df
```

```python
imp = SimpleImputer(strategy='mean')
cleaned = imp.fit_transform(df)
```

```python
cleaned
```

# Remove Duplicates

```python
df2 = pd.DataFrame({'age': [40],'salary': [80000]})
df = pd.concat([df,df2], ignore_index=True)
df
```

```python
df = df.drop_duplicates()
df['age'] = df['age'].astype('int')
```

```python
df
```

# Encoding Categorical Variables

```python
df1 = pd.DataFrame({
    'city': ['Delhi', 'Mumbai', 'Delhi', 'Chennai']
})
df_encoded = pd.get_dummies(df1, columns=['city'])
df_encoded
```

# Scaling Numerical Features

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(df[['age', 'salary']])
scaled
```

# Group BY

```python
data = {
    'Region': ['East', 'West', 'East', 'North', 'West'],
    'Sales': [200, 150, 300, 100, 220],
    'Profit': [20, 15, 30, 10, 25]
}
df = pd.DataFrame(data)
df_group = df.groupby('Region')['Sales'].sum()
print(df_group)
```

# Pivot Table

```python
data = {
    'Region': ['East','East','West','West','North'],
    'Product': ['A','B','A','B','A'],
    'Sales': [200, 250, 300, 100, 120]
}
df = pd.DataFrame(data)
pivot = pd.pivot_table(df, values='Sales', index='Region', columns='Product', aggfunc='sum')
print(pivot)
```
