```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/titanic_dataset.csv')
```

```python
df.head()
```

```python
# Create a countplot
sns.countplot(x='Survived', data=df)
plt.title('Count of Observations per Survived')
plt.show()
```

```python
df['Survived'].value_counts()
```

```python
#x is column
#data is dataframe
#color is mention color name
# you can use another parameter for color saturation like = saturation = 0.1
sns.countplot(x='Pclass', data=df, color="salmon")
plt.show()
```

```python
df['Pclass'].value_counts().plot(kind='bar')
```

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f')
```
