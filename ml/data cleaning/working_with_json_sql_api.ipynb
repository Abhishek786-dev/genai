```python
headers = {
    'x-rapidapi-key': api_key,
    'x-rapidapi-host': host_name
}
```

```python
# get all pages data
df = pd.DataFrame()
for page in range(1,60):
    url = f"/anime?page={page}&size=100&sortBy=ranking&sortOrder=asc"
    conn.request("GET", url, headers=headers)
    res = conn.getresponse()
    data = res.read()
    data = json.loads(data)
    try:
        page_df = pd.DataFrame(data['data'])[['_id', 'title', 'ranking', 'genres', 'episodes']]
        df = pd.concat([df, page_df], ignore_index=True)
        time.sleep(1)
    except Exception as e:
        print(e)
df.shape
```
