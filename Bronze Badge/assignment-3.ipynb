```python
import matplotlib.pyplot as plt

def draw_architecture():
    fig, ax = plt.subplots(figsize=(12, 8))
    
    input_layer = 6
    hidden_layer1 = 5
    hidden_layer2 = 4
    output_layer = 1

    x_position = [0.1, 0.3, 0.5, 0.7]
    
    def draw_circle(x, num_node):
        y_position = [ i /(num_node+1)  for i in range(1, num_node + 1 )]
        for y in y_position:
            circle = plt.Circle((x, y), radius=0.03, color='blue', alpha=0.5)
            ax.add_patch(circle)
        return y_position

    def node_connection(x1, y1_list, x2, y2_list):
        for y1 in y1_list:
            for y2 in y2_list:
                # print([x1, x2], [y1, y2])
                ax.plot([x1, x2], [y1, y2], 'k-', linewidth=0.5)
                
    
    input_result = draw_circle(x_position[0], input_layer)
    hidden1_result = draw_circle(x_position[1], hidden_layer1)
    hidden2_result = draw_circle(x_position[2], hidden_layer2)
    output_result = draw_circle(x_position[3], output_layer)

    node_connection(x_position[0], input_result, x_position[1],  hidden1_result )
    node_connection(x_position[1], hidden1_result, x_position[2],  hidden2_result )
    node_connection(x_position[2], hidden2_result, x_position[3],  output_result )

    input_labels = [
        "TransactionAmount",
        "TransactionTime",
        "MerchantCategory",
        "CustomerAge",
        "AccountBalance",
        "NumTransactionsToday"
    ]
    for i, val in enumerate(input_labels):
        ax.text(0.06, input_result[i] - 0.04, val, fontsize=8, verticalalignment='center')
    
    ax.text(0.68, 0.45, 'Fraud(0/1)', fontsize=8, verticalalignment='center')
draw_architecture()
plt.title('Circle using patches.Circle')
plt.show()
```
