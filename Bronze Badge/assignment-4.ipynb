```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Sample dataset
# Area (sq ft)
area = np.array([800, 1000, 1200, 1500, 1800, 2000, 2200, 2500])

# Number of Bedrooms
bedrooms = np.array([1, 2, 2, 3, 3, 4, 4, 5])

# House Price
price = np.array([150000, 200000, 220000, 300000, 350000, 400000, 420000, 500000])
```

```python
# Combine features
X = np.column_stack((area, bedrooms))
y = price

# Train regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Create mesh grid for surface
area_range = np.linspace(area.min(), area.max(), 20)
bedroom_range = np.linspace(bedrooms.min(), bedrooms.max(), 20)
A, B = np.meshgrid(area_range, bedroom_range)
```

```python
# Predict prices for grid
Z = model.predict(np.c_[A.ravel(), B.ravel()])
Z = Z.reshape(A.shape)
```

```python
# Plot 3D graph
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
# Scatter plot (actual data)
ax.scatter(area, bedrooms, price)

# Regression surface
ax.plot_surface(A, B, Z, alpha=0.5)

ax.set_xlabel("Area (sq ft)")
ax.set_ylabel("Number of Bedrooms")
ax.set_zlabel("House Price")

plt.title("3D Regression Model for House Price Prediction")
plt.show()
```
